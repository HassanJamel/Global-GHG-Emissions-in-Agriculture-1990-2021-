## üå°Ô∏è 4. Strong EDA with Animations

### 4.1 Global Emissions Trend Over Time

```python
# Aggregate global emissions by year
global_yearly = df.groupby('Year')['Value'].sum().reset_index()
global_yearly.columns = ['Year', 'Total_Emissions_kt']

# Calculate year-over-year growth
global_yearly['YoY_Growth_%'] = global_yearly['Total_Emissions_kt'].pct_change() * 100

# Animated Line Chart - Global Emissions Trend
fig = px.line(
    global_yearly, 
    x='Year', 
    y='Total_Emissions_kt',
    title='üåç Global Agricultural Greenhouse Gas Emissions (1990-2021)',
    markers=True,
    labels={'Total_Emissions_kt': 'Total Emissions (kt)', 'Year': 'Year'}
)

fig.update_traces(
    line=dict(width=3, color='#00d4ff'),
    marker=dict(size=8, color='#ff6b6b')
)

fig.update_layout(
    height=500,
    title_font_size=20,
    title_x=0.5,
    xaxis=dict(dtick=2),
    hovermode='x unified'
)

fig.show()
```

### 4.2 Top 10 Emitting Countries - Animated Racing Bar Chart

```python
# Prepare data for racing bar chart
# Filter out aggregate regions and keep only country-level data
exclude_areas = ['World', 'Africa', 'Americas', 'Asia', 'Europe', 'Oceania', 
                 'Eastern Africa', 'Western Africa', 'Northern Africa', 'Southern Africa',
                 'Eastern Asia', 'South-eastern Asia', 'Southern Asia', 'Western Asia',
                 'Central Asia', 'Northern America', 'Central America', 'South America',
                 'Caribbean', 'Eastern Europe', 'Western Europe', 'Northern Europe',
                 'Southern Europe', 'Australia and New Zealand', 'Melanesia', 'Polynesia',
                 'Micronesia', 'European Union', 'Least Developed Countries']

df_countries = df[~df['Area'].isin(exclude_areas)].copy()

# Pivot data for racing bar chart
country_yearly = df_countries.groupby(['Year', 'Area'])['Value'].sum().reset_index()
pivot_df = country_yearly.pivot(index='Year', columns='Area', values='Value').fillna(0)

# Get top 15 countries by total emissions
top_countries = pivot_df.sum().nlargest(15).index.tolist()
pivot_top = pivot_df[top_countries]

print(f"üìä Top 15 Emitting Countries (Total 1990-2021):")
print("-"*50)
for i, country in enumerate(top_countries, 1):
    total = pivot_df[country].sum()
    print(f"{i:2d}. {country}: {total:,.0f} kt")
```

```python
# Animated Bar Chart - Top Countries Over Time
fig = px.bar(
    country_yearly[country_yearly['Area'].isin(top_countries[:10])].sort_values(['Year', 'Value'], ascending=[True, False]),
    x='Value',
    y='Area',
    color='Area',
    animation_frame='Year',
    orientation='h',
    title='üèÜ Top 10 Agricultural GHG Emitters - Racing Bar Chart (1990-2021)',
    labels={'Value': 'Emissions (kt)', 'Area': 'Country'},
    color_discrete_sequence=px.colors.qualitative.Bold
)

fig.update_layout(
    height=600,
    title_font_size=18,
    title_x=0.5,
    showlegend=False,
    yaxis={'categoryorder': 'total ascending'},
    xaxis_title='Total Emissions (kt)',
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        y=-0.15,
        x=0.5,
        xanchor='center',
        buttons=[dict(label='‚ñ∂ Play',
                      method='animate',
                      args=[None, {'frame': {'duration': 500, 'redraw': True},
                                   'fromcurrent': True}])]
    )]
)

fig.show()
```

### 4.3 Interactive Choropleth Map - Global Emissions by Country

```python
# Create animated choropleth map
# Prepare data with ISO codes (simplified mapping)
country_map_data = country_yearly.copy()

fig = px.choropleth(
    country_map_data,
    locations='Area',
    locationmode='country names',
    color='Value',
    hover_name='Area',
    animation_frame='Year',
    color_continuous_scale='Turbo',
    title='üó∫Ô∏è Global Agricultural GHG Emissions by Country (1990-2021)',
    labels={'Value': 'Emissions (kt)'}
)

fig.update_layout(
    height=600,
    title_font_size=18,
    title_x=0.5,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='natural earth',
        bgcolor='rgba(0,0,0,0)'
    )
)

fig.show()
```

### 4.4 Emission Type Analysis (CH4 vs N2O)

```python
# Group by emission type and year
emission_by_type = df.groupby(['Year', 'Element'])['Value'].sum().reset_index()

# Stacked Area Chart
fig = px.area(
    emission_by_type,
    x='Year',
    y='Value',
    color='Element',
    title='üìä Emission Type Breakdown: CH4 vs N2O Over Time',
    labels={'Value': 'Emissions (kt)', 'Element': 'Emission Type'},
    color_discrete_map={
        'Emissions (CH4)': '#ff6b6b',
        'Emissions (N2O)': '#4ecdc4'
    }
)

fig.update_layout(
    height=500,
    title_font_size=18,
    title_x=0.5,
    xaxis=dict(dtick=2),
    hovermode='x unified'
)

fig.show()
```

### 4.5 Emission Sources Analysis (Items)

```python
# Group by emission source (Item)
emission_by_item = df.groupby('Item')['Value'].sum().reset_index()
emission_by_item = emission_by_item.sort_values('Value', ascending=True)

# Horizontal Bar Chart
fig = px.bar(
    emission_by_item,
    x='Value',
    y='Item',
    orientation='h',
    title='üì¶ Emissions by Agricultural Source/Category',
    labels={'Value': 'Total Emissions (kt)', 'Item': 'Source Category'},
    color='Value',
    color_continuous_scale='Viridis'
)

fig.update_layout(
    height=500,
    title_font_size=18,
    title_x=0.5,
    showlegend=False
)

fig.show()
```

## üé® 5. Seaborn Statistical Visualizations

```python
# Set seaborn style
plt.style.use('dark_background')
sns.set_theme(style='darkgrid')

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('üìä Statistical Distribution Analysis', fontsize=20, color='#00d4ff', y=1.02)

# 1. Distribution of Emission Values
ax1 = axes[0, 0]
df_sample = df[df['Value'] > 0]['Value'].sample(min(5000, len(df)))
sns.histplot(np.log10(df_sample + 0.001), bins=50, kde=True, ax=ax1, color='#00d4ff')
ax1.set_title('Distribution of Emission Values (Log Scale)', fontsize=14, color='white')
ax1.set_xlabel('Log10(Emissions)', color='white')
ax1.set_ylabel('Frequency', color='white')

# 2. Box Plot by Emission Type
ax2 = axes[0, 1]
df_box = df[df['Value'] > 0].copy()
df_box['Log_Value'] = np.log10(df_box['Value'] + 0.001)
sns.boxplot(data=df_box, x='Element', y='Log_Value', ax=ax2, palette=['#ff6b6b', '#4ecdc4'])
ax2.set_title('Emission Distribution by Type (Log Scale)', fontsize=14, color='white')
ax2.set_xlabel('Emission Type', color='white')
ax2.set_ylabel('Log10(Emissions)', color='white')

# 3. Yearly Trend Violin Plot
ax3 = axes[1, 0]
years_sample = [1990, 2000, 2010, 2021]
df_violin = df_box[df_box['Year'].isin(years_sample)]
sns.violinplot(data=df_violin, x='Year', y='Log_Value', ax=ax3, palette='viridis')
ax3.set_title('Emission Distribution by Decade', fontsize=14, color='white')
ax3.set_xlabel('Year', color='white')
ax3.set_ylabel('Log10(Emissions)', color='white')

# 4. Correlation Heatmap (Numeric columns)
ax4 = axes[1, 1]
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=ax4, fmt='.2f')
ax4.set_title('Correlation Heatmap', fontsize=14, color='white')

plt.tight_layout()
plt.show()
```

```python
# Calculate KPIs
total_emissions = df['Value'].sum()
avg_emission = df['Value'].mean()
num_countries = df_countries['Area'].nunique()
year_range = f"{df['Year'].min()}-{df['Year'].max()}"
total_records = len(df)

# Latest year stats
latest_data = df[df['Year'] == latest_year]
latest_total = latest_data['Value'].sum()

# First year stats
first_year = df['Year'].min()
first_data = df[df['Year'] == first_year]
first_total = first_data['Value'].sum()

# Growth calculation
total_growth = ((latest_total - first_total) / first_total) * 100

# Top emitter
top_emitter = df_countries.groupby('Area')['Value'].sum().idxmax()
top_emission_value = df_countries.groupby('Area')['Value'].sum().max()
```
